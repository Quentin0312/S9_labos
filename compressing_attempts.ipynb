```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Create a random ndarray with shape (1000, 1220)
original_array = np.random.rand(1000, 1220)
print("original=> ", original_array.shape)

# Apply PCA for dimensionality reduction
target_shape = (100, 122)
pca = PCA(n_components=122)
compressed_array = pca.fit_transform(original_array)
print("compressed => ", compressed_array.shape)

# Inverse transform to get back the original shape
restored_array = pca.inverse_transform(compressed_array)
print("restored => ", restored_array.shape)

# The restored_array will be of shape (100, 1220)
# You can further reduce it to (100, 122) if needed
final_compressed_array = restored_array#[:target_shape[0], :target_shape[1]]

# print(final_compressed_array.shape)

```

```python
import numpy as np

# Create a random array of shape (1000, 1220)
original_array = np.random.rand(1000, 1220)
print(original_array.shape)

# Downsample by averaging blocks of pixels
new_shape = (100, 122)
block_height = original_array.shape[0] // new_shape[0]
block_width = original_array.shape[1] // new_shape[1]

downsampled_array = original_array.reshape(new_shape[0], block_height, new_shape[1], block_width).mean(3).mean(1)

print(downsampled_array.shape)
```
